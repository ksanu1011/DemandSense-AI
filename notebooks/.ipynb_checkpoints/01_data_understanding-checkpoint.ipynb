```python
# Notebook cell (index 0) - simple runnable example
import sys, os
from datetime import datetime

print("Hello from GitHub Copilot — running in Jupyter cell index 0")
print("Time (UTC):", datetime.utcnow().isoformat() + "Z")
print("Python:", sys.version.splitlines()[0])
print("CWD:", os.getcwd())

entries = os.listdir('.')
print("First 10 entries in CWD:")
for name in entries[:10]:
  try:
    size = os.path.getsize(name)
  except OSError:
    size = None
  print(" -", name, f"({size if size is not None else 'unknown'} bytes)")

# Small computation example
n = 100
print("Sum of squares 1..100 =", sum(i * i for i in range(1, n + 1)))
```

```python
# Cell index 1 - reprint results using variables already defined in the notebook
print("Reprinting results from cell 0")
print("Time (UTC):", datetime.utcnow().isoformat() + "Z")
print("Python:", sys.version.splitlines()[0])
print("CWD:", os.getcwd())

print("First 10 entries in CWD:")
for name in entries[:10]:
  try:
    size = os.path.getsize(name)
  except OSError:
    size = None
  print(" -", name, f"({size if size is not None else 'unknown'} bytes)")

if 'n' in globals():
  print("Sum of squares 1..", n, "=", sum(i * i for i in range(1, n + 1)))
else:
  print("Variable 'n' is not defined in the notebook.")
```
