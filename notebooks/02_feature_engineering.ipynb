```python
import pandas as pd

df = pd.read_csv("../data/raw/PJM_Load_hourly.csv")

df["Datetime"] = pd.to_datetime(df["Datetime"])

df = df.sort_values("Datetime")

df = df.set_index("Datetime")

print(df.shape)

df.head()
```

```python
df.shape
```

```python
for lag in range(1, 25):
    df[f"lag_{lag}"] = df["PJM_Load_MW"].shift(lag)

df.head()
```

```python
df.columns
```

```python
len(df.columns)
```

```python
df["rolling_mean_24"] = df["PJM_Load_MW"].rolling(window=24).mean()

df["rolling_std_24"] = df["PJM_Load_MW"].rolling(window=24).std()

df["rolling_min_24"] = df["PJM_Load_MW"].rolling(window=24).min()

df["rolling_max_24"] = df["PJM_Load_MW"].rolling(window=24).max()
```

```python
df.head()
```

```python
print("Before:", df.shape)

df = df.dropna()

print("After:", df.shape)
```
